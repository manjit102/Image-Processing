## Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# <font color = "red">1. Creating Red Image</font>

```python
arr = np.array([[[255, 0 , 0 ], [255, 0 , 0 ], [255, 0 , 0 ],[255, 0 , 0 ]],
                [[255, 0 , 0 ], [255, 0 , 0 ], [255, 0 , 0 ],[255, 0 , 0 ]],
                [[255, 0 , 0 ], [255, 0 , 0 ], [255, 0 , 0 ],[255, 0 , 0 ]],
                [[255, 0 , 0 ], [255, 0 , 0 ], [255, 0 , 0 ],[255, 0 , 0 ]]])

plt.imshow(arr)
plt.show()
```
