# <font color = "brown">Creating Shades of Red - Dense</font>

```python
img = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([i,0,0])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```

```python
img = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0,i,0])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```

# <font color = "brown">Creating Shades of Blue - Dense</font>

```python
img = []
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0,0,i])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```

# <font color = "brown">Merging These Three Dense Shades</font>

```python
img = []

#Red
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([i,0,0])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(256):
        temp.append([i,0,0])
    img.append(temp)
    
#Green
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0,i,0])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(256):
        temp.append([0,i,0])
    img.append(temp)
    
#Blue
for i in range(256):
    temp = []
    for j in range(256):
        temp.append([0,0,i])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(256):
        temp.append([0,0,i])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```

```python
img = []

#Red
for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([i,0,0])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([i,0,0])
    img.append(temp)
    
#Green
for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0,i,0])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([0,i,0])
    img.append(temp)
    
#Blue
for i in range(256):
    temp = []
    for j in range(1536):
        temp.append([0,0,i])
    img.append(temp)
    
for i in range(255,-1,-1):
    temp = []
    for j in range(1536):
        temp.append([0,0,i])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```
