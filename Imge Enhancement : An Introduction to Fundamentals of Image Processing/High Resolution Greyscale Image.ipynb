## 1. Let's Create High Pixel Density Images

```python
image = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(np.random.randint(0,51))
    image.append(temp)
    
image = np.array(image)
plt.imshow(image)
```

```python
image.shape
```

## 2. Exporting the Image

```python
cv.imwrite("random_51.png", image)
```

```python
image = []
for i in range(256):
    temp = []
    for j in range(256):
        
        temp.append(i)
    image.append(temp)
    
image = np.array(image)
plt.imshow(image)
cv.imwrite("image_1.png", image)
```

```python
image = []
for i in range(256):
    temp = []
    for j in range(256):
        
        temp.append(255-i)
    image.append(temp)
    
image = np.array(image)
plt.imshow(image)
cv.imwrite("image_2.png", image)
```
