# <font color ="saffron"> Creating Dense Image</font>

```python
img = []

for i in range(720):
    temp = []
    for j in range(1080):
        temp.append([50, 150, 250])
    img.append(temp)
    
img = np.array(img)
plt.imshow(img)
plt.show()
```
