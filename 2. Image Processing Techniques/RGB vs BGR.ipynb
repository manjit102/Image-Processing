# <font color = "red">RGB vs BGR</font>

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

# <font color = "red">Loading an image using OpenCV</font>

```python
img = cv.imread("rgb.jpg")
```

```python
plt.imshow(img)
```

# <font color = "red">Converting BGR to RGB</font>

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv.imread("Transition.png")
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
plt.imshow(img)
```

```python
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
plt.imshow(img)
```
