# <font color = "red">Frame Extraction</font>

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

# <font color = "red">Loading an Image using OpenCV</font>

```python
img = cv.imread("Transition.png")

img = cv.cvtColor(img, cv.COLOR_RGB2BGR)

plt.imshow(img)
```

```python
plt.imshow(img[ :1000, : , : ])
```

```python
plt.imshow(img[ 500:1000, : , : ])
```

```python
plt.imshow(img[ 1000: , : , : ])
```

```python
r = img[ : , : , 0 ]
g = img[ : , : , 1 ]
b = img[ : , : , 2 ]
```

```python
plt.imshow(r)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
cv.imwrite("Red_t.png",r)
cv.imwrite("Green_t.png",g)
cv.imwrite("Blue_t.png",b)
```
